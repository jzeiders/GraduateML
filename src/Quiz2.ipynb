```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def load_data(file_path):
    return pd.read_csv(file_path, sep='\s+')
   

# Usage
auto_data_df = load_data('/Users/jzeiders/Documents/Code/Learnings/GraduateML/src/auto.data')
```

```python
# Problem 1
def get_auto_regression(df):
    X = df[['displacement']].values
    y = df['mpg'].values

    # Create and fit the model
    model = LinearRegression()
    model.fit(X, y)
    
    return model

print("Slope is ", round(get_auto_regression(auto_data_df).coef_[0],2))
```

```python
# Problem 2
print("r2 is", round(r2_score(auto_data_df['mpg'].values, get_auto_regression(auto_data_df).predict(auto_data_df[['displacement']].values)),2))
```

```python
# Problem 3
# Open question: Do they want normalized?
def get_auto_multi_regression(df):
    X = df[['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin']]
    y = df['mpg']

    # Create and fit the model (without scaling)
    model = LinearRegression()
    model.fit(X, y)

    # Get the coefficient for displacement
    displacement_coef = model.coef_[X.columns.get_loc('displacement')]

    # Round to 2 decimal places
    displacement_coef_rounded = round(displacement_coef, 2)

    print(f"The coefficient associated with displacement is: {displacement_coef_rounded}")

    # Print all coefficients for reference
    for feature, coef in zip(X.columns, model.coef_):
        print(f"{feature}: {round(coef, 2)}")
get_auto_multi_regression(auto_data_df)
```

```python
# Problem 4
# Define the coefficients and intercept
b0 = 50
b1 = 20
b2 = 0.07
b3 = 35
b4 = 0.01
b5 = -10

# Define the input values
gpa = 4.0
iq = 100
gender = 1  # 1 for Female

# Calculate the predicted salary
salary = (
    b0 +  # Intercept
    b1 * gpa +  # GPA
    b2 * iq +  # IQ
    b3 * gender +  # Gender
    b4 * (gpa * iq) +  # Interaction between GPA and IQ
    b5 * (gpa * gender)  # Interaction between GPA and Gender
)

print(f"The predicted salary is ${salary:.2f} thousand")
```

```python
# Problem 5
# For a fixed value of IQ and GPA, females earn more on average than males provided that the GPA is less than 3.5.
# For a fixed value of IQ and GPA, males earn more on average than females provided that the GPA is higher than 3.5.

import plotly.graph_objects as go
import numpy as np

# Model coefficients
b0, b1, b2, b3, b4, b5 = 50, 20, 0.07, 35, 0.01, -10

# Create a grid of GPA and IQ values
gpa = np.linspace(2, 4, 20)
iq = np.linspace(80, 120, 20)
GPA, IQ = np.meshgrid(gpa, iq)

# Calculate salaries for males and females
salary_male = b0 + b1*GPA + b2*IQ + b4*(GPA*IQ)
salary_female = b0 + b1*GPA + b2*IQ + b3*1 + b4*(GPA*IQ) + b5*(GPA*1)

# Create the 3D plot
fig = go.Figure()

# Add surface for males
fig.add_surface(x=GPA, y=IQ, z=salary_male, colorscale='Blues', name='Male')

# Add surface for females
fig.add_surface(x=GPA, y=IQ, z=salary_female, colorscale='Reds', name='Female')

# Update layout
fig.update_layout(
    title='3D Salary Prediction Plot',
    scene=dict(
        xaxis_title='GPA',
        yaxis_title='IQ',
        zaxis_title='Salary (thousands $)',
        xaxis=dict(range=[2, 4]),
        yaxis=dict(range=[80, 120]),
        zaxis=dict(range=[50, 150])
    ),
    width=800,
    height=800,
    autosize=False
)

fig.show()
```
